```python
import os
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt
import pickle


train_data=[]
train_label=[]

flowers=[dirname for dirname in os.listdir('./plants/flower/.crop_data/train_images/') if not dirname.startswith('.')]

for flower_idx,flower in enumerate(flowers):
    images=[image for image in os.listdir('./plants/flower/.crop_data/train_images/'+flowers[flower_idx]+'/') if not image.startswith('.')]
    for image_idx in range(len(images)):
        image=Image.open('./plants/flower/.crop_data/train_images/'+flowers[flower_idx]+'/'+images[image_idx])
        image=image.convert('RGB')
        # W,H=image.size
        # if H<W:
        #     image=image.crop(((W-H)//2,0,(W+H)//2,H))
        # elif H>W:
        #     image=image.crop((0,(H-W)//2,W,(H+W)//2))
        # else:
        #     pass
        image=image.resize((224,224))
        image=np.array(image)
        train_data.append(image)
        train_label.append(flower_idx)

train_data=np.array(train_data)
train_label=np.array(train_label)

with open('train_data_3.pkl','wb') as f:
    pickle.dump(train_data,f)

with open('train_label_3.pkl','wb') as f:
    pickle.dump(train_label,f)
```
