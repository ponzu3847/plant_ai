```python
import requests
import re
from bs4 import BeautifulSoup
import time
from pathlib import Path


category_list=['flower','glass','bulb','herb','garden-tree','fruit-tree','conifer','palm','australia-plants','house-plant',
'succulent-plants','cactus','carnivorous-plant','ferns-bryophytes','orchid','tillandsia','tropical-plants','hydrophyte','vegetables']

plants_dic={}

for category in category_list:
    page_id=1
    plants_dic[category]=set()
    print(category)
    url='https://lovegreen.net/library/type/'+category+'/'

    html=requests.get(url).text
    time.sleep(1.0)
    soup=BeautifulSoup(html,'lxml')
    s_list=soup.select('#postMain > article > div.library-list-ttl.clearfix > h2 > span')

    next_page=['dammy']
    while next_page!=[]:
        # print(s_list)
        for i in range(len(s_list)):
            print(s_list[i].text)
            plants_dic[category].add(s_list[i].text)

        next_page=soup.select('#postMain > div.wp-pagenavi > a.nextpostslink')
        # print(next_page)
        if next_page!=[]:
            page_id+=1
            # next_link=next_page[0].attrs['href']
            next_link='https://lovegreen.net/library/type/'+category+'/page/'+str(page_id)+'/'
            # print('\n'+next_link)
            html=requests.get(next_link).text
            time.sleep(1.0)
            soup=BeautifulSoup(html,'lxml')
            s_list=soup.select('#postMain > article > div.library-list-ttl.clearfix > h2 > span')
    print('\n')

print(plants_dic)
```

```python
import pickle
with open('plants_dic.pkl','wb') as f:
    pickle.dump(plants_dic,f)
```
